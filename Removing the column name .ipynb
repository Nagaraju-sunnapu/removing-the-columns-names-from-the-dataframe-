# Removing the column name 

```python
import numpy as np
import pandas as pd
```

```python
# read the csv file data 
df=pd.read_csv("C:\\Users\\dell\\pandas\\studentmarks1.csv")
print(df)
```

```python
# add new column to the datafram 
df["total"]=df["telugu"]+df["english"]+df["hindi"]+df["maths"]+df["science"]+df["social"]
```

```python
df
```

```python
# remove the total column
df.drop(columns="total")
```

```python
print(df)
```

```python
# df.drop(columns="total") in that we are remove the column called but when we print df this is visible thats why we use inplace=True
```

```python
df.drop(columns="total",inplace=True)
```

```python
# remove the multiple columns 
df.drop(columns=["english","social"])
```

```python
df.drop(columns=["english","social"],inplace=True)
```
